```python
# Import the necessary Python packages

import numpy as np
import scipy.integrate as integrate
from scipy.optimize import fsolve
import matplotlib.pyplot as plt
import math
from tqdm import tqdm
```

```python
#define function 3,4 and 16 part two

#4
def gamma(H,xi):
    return 2*np.pi*R*xi*np.sqrt(2*g*H)/np.sqrt(R**2-xi**2-2*xi**2*np.log(R/xi))

def zeta(H,xi):
    return H*(1-2*xi**2*np.log(R/xi)/(R**2-xi**2))**(-1)

def integral(x,xi):
    return integrate.quad(lambda r: (r**2/x**2-1)*abs(r**2/x**2-1), xi, R)



def x(gamma,f):
    return 1/(2*np.pi)*np.sqrt(gamma/f)
```

```python
R= 0.097
g=9.81
```

```python
Hcalc=0
gammacalc=0
Hres2= list()
fres2= list()

for i in tqdm(range(57,58)):
    xi = R/100*i
    for j in range(1,100):
        H = 0.02+0.04/100*j
        
        
        zetacalc1 = zeta(H,xi)
        gammacalc = gamma(H,xi)
        
        f= 0.0000001
        xcalc = x(gammacalc,f)
        zetacalc2 = integral(xcalc,xi)
        zeta0 =  zetacalc1 -zetacalc2[0]
        zeta0 = round(zeta0,1)

        while zeta0 != 0:
            
            f= f+0.005
            xcalc = x(gammacalc,f)
            zetacalc2 = integral(xcalc,xi)
            zeta0 =  zetacalc1 -zetacalc2[0]
            zeta0 = round(zeta0,1)
        
        fres2.append(f)
        Hres2.append(H)

Hcalc=0
gammacalc=0
Hres3= list()
fres3= list()
for i in tqdm(range(60,61)):
    xi = R/100*i
    for j in range(1,100):
        H = 0.02+0.04/100*j
        
        
        zetacalc1 = zeta(H,xi)
        gammacalc = gamma(H,xi)
        
        f= 0.0000001
        xcalc = x(gammacalc,f)
        zetacalc2 = integral(xcalc,xi)
        zeta0 =  zetacalc1 -zetacalc2[0]
        zeta0 = round(zeta0,1)

        while zeta0 != 0:
            
            f= f+0.005
            xcalc = x(gammacalc,f)
            zetacalc2 = integral(xcalc,xi)
            zeta0 =  zetacalc1 -zetacalc2[0]
            zeta0 = round(zeta0,1)
        
        fres3.append(f)
        Hres3.append(H)

Hcalc=0
gammacalc=0
Hres4= list()
fres4= list()
for i in tqdm(range(64,65)):
    xi = R/100*i
    for j in range(1,100):
        H = 0.02+0.04/100*j
        
        
        zetacalc1 = zeta(H,xi)
        gammacalc = gamma(H,xi)
        
        f= 0.0000001
        xcalc = x(gammacalc,f)
        zetacalc2 = integral(xcalc,xi)
        zeta0 =  zetacalc1 -zetacalc2[0]
        zeta0 = round(zeta0,1)

        while zeta0 != 0:
            
            f= f+0.005
            xcalc = x(gammacalc,f)
            zetacalc2 = integral(xcalc,xi)
            zeta0 =  zetacalc1 -zetacalc2[0]
            zeta0 = round(zeta0,1)
        
        fres4.append(f)
        Hres4.append(H)
```

```python
# Plotting the theoretical curves for beginning of Cat's eye, triangle and square, with experimental values and errors

plt.plot(Hres2,fres2,'r-')
plt.plot(Hres3,fres3,'b-')
plt.plot(Hres4,fres4,'g-')
plt.errorbar([0.04,0.05,0.045,0.03,0.035,0.025,0.0425],[5.5,6.5,6.,4.2,4.8,3.6,5.9],xerr=0.001,fmt='b^',label='Start of triangle state')
plt.errorbar([0.04,0.05,0.045,0.03,0.035,0.025,0.0425],[6.9,7.9,7.3,5.5,6.2,4.8,7.1],xerr=0.001,fmt='g^',label='End of triangle state')
plt.errorbar([0.03,0.035,0.04,0.045,0.05,0.055],[3.5,4.1,4.5,4.9,5.5,6],xerr=0.001,fmt='r.',label='Start of Cat Eye')
plt.errorbar([0.025,0.03,0.035,0.04,0.045,0.05,0.055],[3.45,4.1,4.6,5.3,5.9,6.4,7.],xerr=0.001,fmt='b.', label='End of Cat Eye')
plt.xlabel('H [m]',fontsize=14)
plt.ylabel('f [Hz]',fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
```
